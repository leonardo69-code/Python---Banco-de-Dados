```python
import pandas as pd
```

```python
df = pd.read_csv('https://storage.googleapis.com/aula-06-douglas/dados-brutos/dados.csv')
display(df)
```

```python
df['imc'] = df['peso'] / df['altura'] ** 2
display(df)
```

```python
imcs = df['imc']
imcs = list(imcs)
for i in range(len(imcs)):
  imcs[i] = round(imcs[i],2)
df['imc'] = imcs
display(df)
```

```python
df.to_csv('tratados.csv',index=False)
```
