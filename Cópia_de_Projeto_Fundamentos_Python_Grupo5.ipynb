```python
# Etapa 1 - Adicionar um novo funcionário ao sistema, solicitando informações como nome, cargo, salário, e-mail, etc.
# Etapa 2- Atualizar as informações de um funcionário existente com base em um identificador único (como um número de identificação ou nome). Exibir informações de um funcionário específico.
# Etapa 3 - Listar todos os funcionários cadastrados.
# Etapa 4 - Permitir a remoção de um funcionário do sistema.

# Observações Finais: Certifique-se de implementar um sistema de identificação único para cada funcionário.
# Valide as entradas do usuário para garantir que as operações sejam realizadas corretamente e sem erros.
# Documente adequadamente seu código com comentários explicativos para ajudar na compreensão.
# Esse projeto permite aos alunos praticar e consolidar conceitos fundamentais em Python, como operadores, estruturas de controle,
# estruturas de dados e funções, enquanto constroem um aplicativo útil para gerenciar informações de funcionários de uma empresa fictícia.
```

```python
# Inicializando o dicionário para armazenar os funcionários
funcionarios = {}
id_counter = 1  # Contador para fornecer IDs únicos aos funcionários

# Função para adicionar um novo funcionário
def adicionar_funcionario():
    global id_counter
    nome = input("Digite o nome do funcionário: ")
    cidade = input("Digite a cidade do funcionário: ")
    estado = input("Digite o estado do funcionário: ")
    idade = int(input("Digite a idade do funcionário: "))
    escolaridade = input("Digite a escolaridade do funcionário: ")
    cargo = input("Digite o cargo do funcionário: ")
    salario = float(input("Digite o salário do funcionário: "))
    email = input("Digite o e-mail do funcionário: ")

    # Criando um dicionário para armazenar os dados do funcionário
    funcionario = {
        "nome": nome,
        "cidade": cidade,
        "estado": estado,
        "idade": idade,
        "escolaridade": escolaridade,
        "cargo": cargo,
        "salario": salario,
        "email": email
    }

    # Adicionando o funcionário ao dicionário geral com um ID único
    funcionarios[id_counter] = funcionario
    print(f"\nFuncionário {nome} adicionado com sucesso! ID: {id_counter}")
    id_counter += 1

# Função para atualizar as informações de um funcionário
def atualizar_funcionario():
    id_func = int(input("Digite o ID do funcionário que deseja atualizar: "))
    if id_func in funcionarios:
        nome = input("Digite o novo nome do funcionário: ")
        cidade = input("Digite a nova cidade do funcionário: ")
        estado = input("Digite o novo estado do funcionário: ")
        idade = int(input("Digite a nova idade do funcionário: "))
        escolaridade = input("Digite a nova escolaridade do funcionário: ")
        cargo = input("Digite o novo cargo do funcionário: ")
        salario = float(input("Digite o novo salário do funcionário: "))
        email = input("Digite o novo e-mail do funcionário: ")

        # Atualizando os dados do funcionário
        funcionarios[id_func] = {
            "nome": nome,
            "cidade": cidade,
            "estado": estado,
            "idade": idade,
            "escolaridade": escolaridade,
            "cargo": cargo,
            "salario": salario,
            "email": email
        }
        print(f"\nFuncionário de ID {id_func} atualizado com sucesso!")
    else:
        print("\nFuncionário não encontrado.")

# Função para listar todos os funcionários
def listar_funcionarios():
    if funcionarios:
        print("\nLista de funcionários:")
        for id_func, dados in funcionarios.items():
            print(f"\nID: {id_func}, Nome: {dados['nome']}, Cargo: {dados['cargo']}, Salário: {dados['salario']:.2f}")
    else:
        print("\nNenhum funcionário cadastrado.")

# Função para remover um funcionário do sistema
def remover_funcionario():
    id_func = int(input("Digite o ID do funcionário que deseja remover: "))
    if id_func in funcionarios:
        del funcionarios[id_func]
        print(f"\nFuncionário de ID {id_func} removido com sucesso!")
    else:
        print("\nFuncionário não encontrado.")

# Função principal para exibir o menu e controlar as operações
def menu():
    while True:
        print("\nSistema de Gerenciamento de Funcionários")
        print("1 - Adicionar novo funcionário")
        print("2 - Atualizar funcionário")
        print("3 - Listar todos os funcionários")
        print("4 - Remover funcionário")
        print("5 - Sair")

        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            adicionar_funcionario()
        elif opcao == "2":
            atualizar_funcionario()
        elif opcao == "3":
            listar_funcionarios()
        elif opcao == "4":
            remover_funcionario()
        elif opcao == "5":
            print("Encerrando o sistema...")
            break
        else:
            print("Opção inválida. Tente novamente.")

# Iniciando o sistema
menu()
```
