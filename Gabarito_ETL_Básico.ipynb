```python
import pandas as pd
```

```python
df = pd.read_csv('https://storage.googleapis.com/etl-basico/dados-brutos/dados_brutos.csv')
display(df)
```

```python
# Transformar cada coluna do df em uma lista
nomes = list(df['nome'])
idades = list(df['idade'])
salarios = list(df['sal√°rio'])
cargos = list(df['cargo'])
departamentos = list(df['departamento'])
```

```python
# Fazendo o tratamento para excluir os registros com nan , erro e Desconhecido
nome = []
idade = []
salario = []
cargo = []
departamento = []
for i in range(len(nomes)):
  if (str(nomes[i]).strip().lower()not in['nan', 'erro', 'desconhecido'] and
      str(idades[i]).strip().lower()not in['nan', 'erro', 'desconhecido'] and
      str(salarios[i]).strip().lower()not in['nan', 'erro', 'desconhecido'] and
      str(cargos[i]).strip().lower()not in['nan', 'erro', 'desconhecido'] and
      str(departamentos[i]).strip().lower()not in['nan','erro','desconhecido']):
    nome.append(nomes[i])
    idade.append(idades[i])
    salario.append(salarios[i])
    cargo.append(cargos[i])
    departamento.append(departamentos[i])
```

```python
# Convertendo as listas tratadas em um df pandas

df = pd.DataFrame(zip(nome, idade, salario, cargo, departamento),
                  columns=['nome', 'idade', 'salario', 'cargo', 'departamento'])
display(df)
```

```python
# Convertendo as listas tratadas em um df pandas
df = pd.DataFrame({
    'nome':nome,
    'idade':idade,
    'salario':salario,
    'cargo':cargo,
    'departamento':departamento
})
display(df)
```
