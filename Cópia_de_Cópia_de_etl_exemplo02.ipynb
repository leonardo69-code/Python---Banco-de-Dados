```python
import pandas as pd
import gcsfs
from google.colab import auth
from google.cloud import storage
```

```python
df = pd.read_csv('https://storage.googleapis.com/projeto-bucket-leonardo/dados-brutos/dados_inconsistentes%20-%20dados_inconsistentes.csv.csv')
display(df)
```

```python
auth.authenticate_user()
```

```python
client = storage.Client()
bucket = client.bucket('projeto-bucket-leonardo')
```

```python
print(df.columns)
```

```python
nome = df['nome'].tolist() if 'nome' in df.columns else []
idade = df['idade'].tolist() if 'idade' in df.columns else []
salario = df['salário'].tolist() if 'salario' in df.columns else []
cargo = df['cargo'].tolist() if 'cargo' in df.columns else []
departamento = df['departamento'].tolist() if 'departamento' in df.columns else []
```

```python
print(len(nome), len(idade), len(salario), len(cargo), len(departamento))
```

```python
min_len = min(len(nome), len(idade), len(salario), len(cargo), len(departamento))

nome = nome[:min_len]
idade = idade[:min_len]
salario = salario[:min_len]
cargo = cargo[:min_len]
departamento = departamento[:min_len]
```

```python
df_tratado = pd.DataFrame({
    'nome': nome,
    'idade': idade,
    'salario': salario,
    'cargo': cargo,
    'departamento': departamento
})
```

```python
import os
print(os.listdir('/tmp'))
```

```python
print(df_tratado.head())
```

```python
df_tratado.to_csv('/tmp/dados_tratados.csv', index=False)
```

```python
blob = bucket.blob('dados-tratados/dados_tratados.csv')
blob.upload_from_filename('/tmp/dados_tratados.csv')
print("Arquivo tratado enviado para 'dados-tratados' com sucesso.")
```
